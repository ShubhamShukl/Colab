# Removind duplicate entries

```python
import pandas as pd

# Load the CSV file into a DataFrame
df1 = pd.read_csv('/content/tweets_BoycottNetflixIndia.csv')

# Remove duplicate rows based on the 'text' column
df1_cleaned = df1.drop_duplicates(subset='text')

# Save the cleaned DataFrame back to a CSV file
df1_cleaned.to_csv('tweets_BoycottNetflixIndia_Cleaned.csv', index=False)

print("Duplicate entries removed and DataFrame saved to cleaned_file.csv.")
```

# Dataset Labelling

```python
import pandas as pd

# Example dataframe for illustration
df=pd.read_csv('/content/tweets_BoycottNetflixIndia_Cleaned.csv')

# Add a new column 'label' with default None
df['label'] = None




# Iterate through each row in the dataframe
for i in range(len(df)):
    # Display the current text entry
    print(f"Text: {df.loc[i, 'text']}")

    # Ask the user to label the text
    user_label = input(f"Please provide a label for the above text: ")

    # Add the user input to the label column
    df.loc[i, 'label'] = user_label

    # Optionally display the updated row
    print(f"Row updated: {df.loc[i]}")
    print("-" * 40)

# Once labeling is done, ask the user for the file name to save the dataframe
file_name = input("Enter the file name to save the DataFrame (without extension): ")

# Save the dataframe as a CSV file
df.to_csv(f"{file_name}.csv", index=False)

print(f"DataFrame saved as {file_name}.csv")
```
